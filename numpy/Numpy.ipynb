###  graficare i dati monodimensionali ###
### possiamo usare matplotilab per creare dei grafici di funzioni ###

```python
#Creare dei dati per x e y= x^2
x= np.arange(-9.,10.,1)
y=x**2

print(x)
print(y)
```

```python
#facciomo il grafico della funzione
plt.plot(x,y) #in questo modo creiamo il grafico

#mostriamo il grafico creato
plt.show()
```

```python
y2 = x**3
```

```python
#mettiamo due grafici nella stessa immagine

plt.plot(x,y, label="x=x^2")
plt.plot(x,y2,label="y=x^3")
plt.legend() #mostra una legenda

#mostriamo il grafico creato e la rimuove dalla memoria
plt.show()
```

### come visualizzare le matrice che Ã¨ anche lo stesso modo di visualizzare le immagini ###
### mostrare array bidimensionale come immagini(heatmaps)###

```python
# creiamo una matrice, ovvero un array bidimensionale.
m1= np.random.randn(10,10)
print(m1)
```

```python
plt.imshow(m1)
plt.show()
```

```python
# creiamo una matrice piu interessante
data = []

for i in range(10): #i va da 0 a 9
    data.append(np.arange(i,i+11)) # aggiungi (append) alla fine
    
matrix = np.array(data)

print(matrix)

plt.imshow(matrix)
plt.show()
```

### indicizzazione degli elementi di una matrice ###

```python
# in una dim
vettore= np.arange(10,21)
print(vettore)

print(vettore [1])
print(vettore [4])
```

```python
#per le matrici abbiamo bisogno di indirizzare gli elementi con due diversi indici come faccioamo per individuare gli elementi di una matrice in algebra
print(matrix)

print('elemento 3,5')
print(matrix[2,4]) #riga,colonna
```

```python
# selezionare una sottomatrice (slice)

print(matrix)
print()
#si indicizza con le quadre
print(matrix[1:6, 1:6]) #1 compreso, 6 escluso ([righe,colonne])
```

```python
print(matrix[5:, 5:]) # questo Ã¨ uno slicing
```

### le slice sottomatrici condividono la memoria con la matrice originale###

```python
sottomatrice= matrix[:5, :5]

print(matrix)
print(sottomatrice)
```

```python
sottomatrice [:] = 0 # questa dicitura prende tutte le righe e tutte le colonne
```

### indicizzare con maschera di booleani###
Ã¨ possibile indicizzare gli elementi di una matrice che rispettano una certa condizione utilizzando una matrice di booleani della stessa forma della matrice da indirizzare.  La matrice di booleani viene chiamata maschera booleana.

```python
# vogliamo aumentare di 100 tutti i valori pari contenuti nella matrice
```

```python
mask = (matrix % 2 == 0)
print (matrix)
print(mask)
```

```python
print (matrix[mask]) # sovrappone maschera alla matrice (per questo devono avere la stessa dimensione)
```

```python
matrix[mask] = matrix [mask]+100

# cOSA SUCCEDE NELLA RIGA QUI SOPRA?
#1. MATRIX[MASK]= PRENDI DALLA MATRICE MATRIX LE POSIZIONI RELATIVE AI TRUE DENTRO MASK E SETTA I VALORI IN QUELLE POSIZIONI A ....
#2. MATRIX [MASK]+100
# ... I VALORI PRESI DALLA MATRICE MATRIX LE POSIZIONI RELATIVE AI TRUE DENTRO MASK ED AUMENTATI DI 100
print(matrix)
```
